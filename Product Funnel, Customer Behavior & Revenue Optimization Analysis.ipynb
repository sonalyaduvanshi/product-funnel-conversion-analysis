```python
import pandas as pd
df = pd.read_excel( "Product Funnel, Customer Behavior & Revenue Optimization Analysis.xlsx")
```

```python
df
```

```python
df.head()
df
```

```python
df = df.copy()
df
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df
```

```python
df = df[df["Monetary"] >= 0]
df
```

```python
df = df[df["Session_Count"] > 0]
df
```

```python
df.loc[:, "High_Value_Customer"] = df["Monetary"].apply(
    lambda x: "Yes" if x > 2000 else "No"
)
df
```

```python
df.loc[:, "Engagement_Level"] = df["Session_Count"].apply(
    lambda x: "High" if x > 40 else "Medium" if x > 20 else "Low"
)
df
```

```python
df.to_csv("cleaned_ecommerce_data.csv", index=False)
df
```
